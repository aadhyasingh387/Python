Scatter Plot

```python
import matplotlib.pyplot as plt 
day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
c = ["r" , "y" , "g" , "b" , "r" , "g" , "r"]
s = [100, 20,40,50,20,100,70]
plt.scatter (day, no , color = c , sizes =s , alpha = 0.5 , marker = "*" , edgecolor = "g" , linewidths=4 )
plt.title("Scatter Plot" , fontsize=15)
plt.xlabel("Day" , fontsize=15)
plt.ylabel("No." , fontsize=15)
plt.show()
```
